```python
%matplotlib nbagg  
#일반적으로 맨위에 설정한후 진행함

import numpy as np 
import matplotlib
import matplotlib.pyplot as plt #그림그릴때 plt 이용

#matplotlib의 기본설정   plt.rcParams[]
#폰트같은것에 문제가 발생하는데 그걸 막기위해 기본설정

plt.rcParams["axes.grid"]=False #그래프안에 격자를 표현해줄거냐 말거냐 TRUE 격자출력
plt.rcParams["figure.figsize"]=(5,6) #가로길이 세로길이


x_data=range(50) #가로축 데이터
y_data=np.random.randint(-100,100,50).cumsum()

plt.title("가격변동 추이")
plt.xlabel("기간(일)")
print(x_data)
plt.plot(x_data , y_data, "r")
plt.show()
```
