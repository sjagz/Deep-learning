```python
#numpy array 연결(2개의 array를 행,열,방향으로 연결)

import numpy as np

arr1=np.array([[1,2,3],
               [4,5,6],
              ])
arr2=np.array([[1,1,1],
               [2,2,2],
               [3,3,3],
               [4,4,4]
              ])
print(arr1)
print(arr2)

#행은 행 열은 열끼리 숫자가 같으면 붙게됨


result=np.vstack((arr1,arr2)) #결과로 새로운 배열이 리턴 열방향 수직방향의 리턴이야 라는 뜻a

print(result)



```
