```python
#1.사용자가 평가한 모든 영화의 전체 평점

import numpy as np
import pandas as pd

#df = pandas.read_csv('filename.csv', encoding='utf-8')


ratings = pd.read_csv('./data/movie/ratings.csv')
movie = pd.read_csv('./data/movie/movies.csv' )
cols=["각 사용자가 평가한 모든 영화의 전체평점","key"]
i=1
df1 = pd.DataFrame(columns=cols, index=range(1))
display(df1)
print("#1.사용자가 평가한 모든 영화의 전체 평점")
print(ratings["rating"].mean())
df1.iloc[0,0]=ratings["rating"].mean()
df1.iloc[0,1]=1
display(df1)
```

```python
print("#2.각 사용자별 평균평점")
cols2 = ["각사용자별평균평점", 'c2']
df2 = pd.DataFrame(columns=cols2, index=range(len(ratings["userId"].unique())+1))


for i in ratings["userId"].unique():
    df2.iloc[i,0]=ratings[ratings["movieId"]==i]["rating"].mean()
    df2.iloc[i,1]=i
display(df2)
```

```python
print("3번 문제")


cols = ["movieId","각 영화별 평균평점","key"]
c1=len(ratings["movieId"].unique())

df3 = pd.DataFrame(columns=cols, index=range(c1+1))


i=0
for j in ratings["movieId"].unique():
    
    df3.iloc[i,0]=j
    df3.iloc[i,1]=ratings[ratings["movieId"]==j]["rating"].mean()
    df3.iloc[i,2]=i
    i=i+1
display(df3)

```

```python
print("4번 문제")

df4=df3
df4["가장 높은평점"]=np.nan
df4["가장평점이 높은 영화"]=np.nan

func = lambda x : x.max()  
#display(df4[["각 영화별 평균평점"]].apply(func,axis=0).values)
print(df4[["각 영화별 평균평점"]].apply(func,axis=0))

df4.loc[0,["가장 높은평점"]]=df4[["각 영화별 평균평점"]].apply(func,axis=0).values

i=0

for k in df4["movieId"][df3["각 영화별 평균평점"] == df4.loc[0,"가장 높은평점"]] :
    df4.loc[i,"가장평점이 높은 영화"] = k
    i=i+1
display(df4)
#print(df3["각 영화별 평균평점"] == df.loc["가장높은평점"])
```
