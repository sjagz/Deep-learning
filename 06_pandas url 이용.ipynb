```python
#조인 방식 데이터프레임 -> merge
#4가지 아웃 인 레프트 라이트 sql의 조인과 같다

#inner join  (2개의 테이블 기준의 중복된 값)키값이 공통적인것만 뽑아냄

#full outer join (2개의 테이블 기준의 모든값)키값에 해당하는것 다 뽑음

#left outer join (full outer join에서 왼쪽에 해당하는 키값만 출력 왼쪽에 해당하지않는 값은 날림)

#right outer join((full outer join에서 오른쪽에 해당하는 키값만 출력 오른쪽에 해당하지않는 값은 날림))

import numpy as np
import pandas as pd

df1=pd.DataFrame({"학번":["201901","201902","201903"],
                  "이름":["이지안","박동훈","홍길동"]
                })
df2=pd.DataFrame({"학번":["201901","201902","201904"],
                  "학과":["정통","컴퓨터","경영"]
                })

pd.merge(df1,df2,on="학번", how="inner")
pd.merge(df1,df2,on="학번", how="outer")
pd.merge(df1,df2,on="학번", how="left")
pd.merge(df1,df2,on="학번", how="right")






```
