```python
#py mysql module 을 이용

import pymysql.cursors    #conda install pymysql 설치해야함 아나콘다 콘솔창에 
import numpy as np
import pandas as pd



# mysql 포트번호 를 알아야함 아이디와 패스워드 -> 여러 데이터베이스중 어떤 데이터베이스에 접근할지 지정

conn=pymysql.connect(host ="localhost",#mysql이 설치된 컴퓨터 ip 
                user="python",    # 사용자 계정 ID
                password="python", # 사용자 password
                db="library",     # 사용할 database 
                charset="UTF8"
                ) # 출력화면   <pymysql.connections.Connection at 0x6942dd8>

#sql을 작성 
#제공한 script 파일에 books 라는 table이 있다 


keyword =input("키워드 입력하세요")
sql = "SELECT bisbn,btitle,bauthor,bprice FROM books WHERE btitle LIKE '%{}%'".format(keyword)



df=pd.read_sql(sql, con=conn) #pandas 가 sql을 실행시켜서 그결과를 dataframe으로 생성해주는 함수
#df.to_json("./data/sample/books.json")

display(df)
conn.close()#종료처리







```
